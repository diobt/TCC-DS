```python
import pandas as pd
import matplotlib. pyplot as plt
import sqlite3
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2020_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';' )   #read the csv file (put 'r' before the path string to address any special characters in the path, such as '\'). Don't forget to put the file name at the end of the path + ".csv"
df2 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2020_Parte2_14fev2022.csv', on_bad_lines='skip', sep=';')
df3 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2021_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';' )
df4 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2021_Parte2_14fev2022.csv', on_bad_lines='skip', sep=';')
df5 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2022_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';')
dfbrasil = pd.concat([df,df2,df3,df4,df5])
```

```python
#criação banco de dados
conn= sqlite3.connect('brasilcovid.db')
```

```python
#criação tabela
dfbrasil.to_sql('brcov',conn)
```

```python
#codigo da figura 19
query1="""
select regiao,estado ,municipio, casosAcumulado, populacaoTCU2019, data from brcov where data in ('2020-07-31', '2020-12-31' , '2021-06-30' , '2022-02-14') and regiao ='Brasil' ;
"""

grafAno=pd.read_sql(query1,conn)


fig= sns.catplot(data = grafAno,x='data',y='casosAcumulado', kind='bar', aspect=50/25)
sns.lineplot(data=grafAno,x='data',y='casosAcumulado',color ='black')
plt.show(fig)
```

```python
#codigo figura 22
query1="""
select regiao,municipio, casosAcumulado,populacaoTCU2019,data from brcov where municipio in ('Porto Velho',
'Manaus',
'Rio Branco',
'Campo Grande',
'Macapá',
'Brasília',
'Boa Vista',
'Cuiabá',
'Palmas',
'São Paulo',
'Teresina',
'Rio de Janeiro',
'Belém',
'Goiânia',
'Salvador',
'Florianópolis',
'São Luís',
'Maceió',
'Porto Alegre',
'Curitiba',
'Belo Horizonte',
'Fortaleza',
'Recife',
'João Pessoa',
'Aracaju',
'Natal',
'Vitória' );
"""
pCem=pd.read_sql(query1,conn)



pCem['casosPorCemMil'] =( pCem['casosAcumulado'] / pCem['populacaoTCU2019'] )* 100000




plt.figure(figsize=(30,15))
fig=sns.barplot(data=pCem,x='municipio',y='casosPorCemMil', ci=False)
plt.title('total de casos por 100mil habitantes no Brasil', fontsize=20)
plt.xlabel('capitais', fontsize=20)
plt.ylabel('casos por 100mil habitantes', fontsize=20)
plt.xticks(rotation= 90, fontsize=20)
plt.show(fig)
```
