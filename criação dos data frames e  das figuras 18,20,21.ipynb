```python
import pandas as pd
import matplotlib. pyplot as plt
import sqlite3
import seaborn as sns
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2020_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';', engine = 'python')   #read the csv file (put 'r' before the path string to address any special characters in the path, such as '\'). Don't forget to put the file name at the end of the path + ".csv"
df2 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2020_Parte2_14fev2022.csv', on_bad_lines='skip', sep=';', engine = 'python')
df3 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2021_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';', engine = 'python')
df4 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2021_Parte2_14fev2022.csv', on_bad_lines='skip', sep=';', engine = 'python')
df5 =pd.read_csv (r'/content/drive/MyDrive/TCC.2022/DATASET/HIST_PAINEL_COVIDBR_2022_Parte1_14fev2022.csv', on_bad_lines='skip', sep=';', engine = 'python')
```

```python
dfbrasil201=df[df['regiao']=='Brasil']
dfbrasil202=df2[df2['regiao']=='Brasil']
dfbrasil211=df3[df3['regiao']=='Brasil']
dfbrasil212=df4[df4['regiao']=='Brasil']
dfbrasil221=df5[df5['regiao']=='Brasil']
dfbrasil= pd.concat([dfbrasil201,dfbrasil202,dfbrasil211,dfbrasil212,dfbrasil221], axis=0, ignore_index= True)
```

```python
#codigo grafico do crescimento de casos de covid fig 18
x=dfbrasil.data
y=dfbrasil.casosAcumulado
plt.figure(figsize=(30,15))
plt.plot(x,y)
plt.title('Crescimento dos casos de COVID-19 no Brasil', fontsize=20)
plt.xlabel('data', fontsize=20)
plt.ylabel('quantidade de casos x 10^7', fontsize=20)
plt.xticks(['2020-02-25','2020-04-30','2020-06-30','2020-08-30','2020-10-30','2020-12-30','2021-02-28','2021-04-30','2021-06-30','2021-08-30','2021-10-30','2021-12-31','2022-02-14'],rotation = 45, fontsize= 20)
plt.yticks([0,5000000,10000000,15000000,20000000,25000000,30000000], fontsize= 20)
plt.show()
```

```python
#codigo da taxa de crescimento de novos casos fig 20
x=dfbrasil.data
y=dfbrasil.casosNovos
plt.figure(figsize=(30,15))
plt.plot(x,y)
plt.title('Novos infectados comparados ao dia anterior de 2020 ate 2022', fontsize= 20)
plt.ylabel('numero de novos infectados', fontsize=20)
plt.xlabel('data', fontsize= 20)
plt.xticks(['2020-02-25','2020-04-30','2020-06-30','2020-08-30','2020-10-30','2020-12-30','2021-02-28','2021-04-30','2021-06-30','2021-08-30','2021-10-30','2021-12-31','2022-02-14'],rotation = 45, fontsize= 20)
```
